```python
#LIST OF MOST POPULAR GENRES
```

```python
sns.countplot(x='Genre',data=df)
plt.title("List of Genres")
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
rating=df['Film'].value_counts(10).head(10)
sns.barplot(x=rating.values,y=rating.index,palette='magma')
plt.title('Highly Rated Movie')
plt.xlabel('Rating')
plt.ylabel('Film')
plt.show()
```

```python
#HIGHLY AUDIENCE RATED MOVIES BASED ON GENRES 
```

```python
sns.barplot(x='Genre',y='Audience Ratings %',data=df,palette='magma')
plt.title('Highly rated movie based on genres')
plt.xlabel('Genre')
plt.ylabel('Audience Rating')
plt.show()
```
