```python
#MOST WIINING TEAMS
```

```python
top_teams=matches['winner'].value_counts().head(10)
sns.barplot(x=top_teams.values,y=top_teams.index,palette='viridis')
plt.xlabel('Teams')
plt.ylabel('Wins')
plt.show()
```

```python
#MOST SUCCESSFUL CAPTAINS
```

```python
captains=matches['player_of_match'].value_counts().head(10)
sns.barplot(x=captains.values,y=captains.index,palette='magma')
plt.xlabel('Top 10 players of MAN OF MATCH AWARDS')
plt.ylabel('Awards')
plt.show()
```

```python
toss_match_win=matches[matches['toss_winner']==matches['winner']]
toss_win_per=len(toss_match_win)/len(matches)*100
print('Toss winner also match winner in',toss_win_per,'chance')
```

```python
sns.countplot(data=matches,x='season',palette='coolwarm')
plt.title('Number of matches in season')
plt.xticks(rotation=45)
plt.show()
```

```python
stadiums=matches['venue'].value_counts().head(10)
sns.barplot(x=stadiums.values,y=stadiums.index,palette='plasma')
plt.title('Stadiums')
plt.show()
```
